```python
import sys
import numpy as np
import pdb
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
df_all_melted=pd.read_csv('../data_inhouse_fmri/data_for_fig_allsnr.csv')
df_raw_melted=pd.read_csv('../data_inhouse_fmri/data_for_fig_rawsnr.csv')
y_reliabilities = np.load('../data_inhouse_fmri/data_for_fig_reliability.npz.npy')
```

```python
regions = df_raw_melted.region.unique() 
regions_keep = ['IFGop_L','IFGop_R'] # can choose a subset
regions_keep = ['Frontal Cortex','Visual Cortex'] # can choose a subset
#regions_keep = df_raw_melted.region.unique() # or keep all 

index = [r in regions_keep for r in df_raw_melted.region]


avgname = ''
avgname='FO'

# plot parameters 
sns.set_style('white')
params = {
   'axes.labelsize': 10,
   'axes.titlesize': 8,
    'axes.titleweight': 'normal',
    'axes.labelweight':'normal',
   'text.fontsize': 8,
   'legend.fontsize': 8,
   'xtick.labelsize': 8,
   'ytick.labelsize': 8,
   'text.usetex': False,
    'text.color':'black'
    
}

plt.rcParams.update(params)
plt.rcParams['font.sans-serif']='Helvetica'
plt.rcParams['font.family']='sans-serif'

fig,axes = plt.subplots(1,3,figsize=(6,4),sharex=True)

sns.despine()


# PLOT 1 
snrtype = 'raw SNR'
subs = df_raw_melted.subject.unique()
df_for_plot = df_raw_melted.loc[index,] 
regions = df_for_plot.region.unique()[::-1]
ys = np.empty((len(subs),len(regions)))
for si,sub in enumerate(subs):
    x=0
    xs=np.array([])
    for region in regions:
            xs = np.append(xs,x)
            y = df_for_plot.loc[(df_for_plot.subject==sub)&
                        (df_for_plot.region==region),snrtype].as_matrix()
            ys[si,x] = y
            x+=1
    axes[0].plot(xs,ys[si,:],c='k',linewidth=0.3)
    
ys_raw = ys.copy() 
axes[0].plot(xs,np.mean(ys,axis=0),c='b',linewidth=2,label='group mean')
axes[0].errorbar(xs,np.mean(ys,axis=0),yerr=np.std(ys,axis=0)/np.sqrt(len(ys)),c='b')
axes[0].set_xticks(xs)
axes[0].set_xticklabels(regions,rotation=90)
#axes[0].set_xlabel('Region')
axes[0].set_xlim([-0.2,len(regions)-0.8])
axes[0].set_ylabel('Raw SNR')


# PLOT 2
snrtype = 'all v fix SNR'
subs = df_all_melted.subject.unique()
df_for_plot = df_all_melted.loc[index,]
regions = df_for_plot.region.unique()[::-1]
ys = np.empty((len(subs),len(regions)))
for si,sub in enumerate(subs):
    x=0
    xs=np.array([])
    for region in regions:
            xs = np.append(xs,x)
            y = df_for_plot.loc[(df_for_plot.subject==sub)&
                        (df_for_plot.region==region),snrtype].as_matrix()
            ys[si,x] = y
            x+=1
    axes[1].plot(xs,ys[si,:],c='k',linewidth=0.3)
ys_all = ys.copy() 
axes[1].plot(xs,np.mean(ys,axis=0),c='b',linewidth=2,label='group mean')
axes[1].errorbar(xs,np.mean(ys,axis=0),yerr=np.std(ys,axis=0)/np.sqrt(len(ys)),c='b')
axes[1].set_xticks(xs)
axes[1].set_xticklabels(regions,rotation=90)
axes[1].set_xlabel('Region')
axes[1].set_xlim([-0.2,len(regions)-0.8])
axes[1].set_ylabel('Functional SNR (avg T-statistic)')
    
    
## Plot 3
for sub in range(y_reliabilities.shape[1]):
    if sub!=y_reliabilities.shape[1]-1:
        axes[2].plot([0,1],y_reliabilities[:,sub],c='k',linewidth=0.3)
    else:
        axes[2].plot([0,1],y_reliabilities[:,sub],c='k',linewidth=0.3,label='individual subject')

axes[2].plot([0,1],np.mean(y_reliabilities,axis=1),c='b',linewidth=2,label='group mean')
axes[2].errorbar([0,1],np.mean(y_reliabilities,axis=1),yerr=np.std(y_reliabilities,axis=1)/np.sqrt(y_reliabilities.shape[1]),c='b')
axes[2].set_xticks([0,1])
axes[2].set_xticklabels(['Frontal Cortex','Visual Cortex'],rotation=90)
axes[2].set_xlim([-0.2,1.2])
axes[2].set_ylabel('Reliability (Split-half)')
#axes[2].set_xlabel('Region')

    
plt.legend(bbox_to_anchor=(.8, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.savefig('../Figures/Supp_SNR_Reliability.png',dpi=400,bbox_inches="tight")
plt.show()
```
