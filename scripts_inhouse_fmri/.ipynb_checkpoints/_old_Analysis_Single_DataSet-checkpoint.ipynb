```r
get_largedataset<-function(foldername){


folder = paste(datadir,foldername,sep='')
#remove(subdata)
subdata <-data.frame(sid=0, cvm= 0, cvsd=0,roi='A',cvk=0,permmean=0,perm95=0,perm5=0,cor1=0,cor2=0,err1=0,err2=0,pvalue_perm=0,ruletype=0,stringsAsFactors=FALSE)
i <-1 # subject counter
ii<-1 # row counter 
Fsubs = c()
Osubs = c()
Fsubsnull = matrix(,21,1000)
Osubsnull = matrix(,21,1000)

for (s in subids){

    temp_dataset<- read.table(paste0(folder,"roiFrontal_Lobesubfmri_",s, "_TRInulldist.txt"),sep='\t')
    Fsubsnull[i,]<- as.vector(t(temp_dataset))
    temp_dataset2<- read.table(paste0(folder,"roiOccipital_Lobesubfmri_",s, "_TRInulldist.txt"))
    Osubsnull[i,]<- as.vector(t(temp_dataset2))
    Fsub <-read.table(paste0(folder,"roiFrontal_Lobesubfmri_",s, "_TRI.txt"))
    Fsubs[i] <-Fsub
    Osub <-read.table(paste0(folder,"roiOccipital_Lobesubfmri_",s, "_TRI.txt"))
    Osubs[i] <-Osub
    Fmat<-read.table(paste0(folder,"roiFrontal_Lobesubfmri_",s, "_TRImatrix.txt"))
    Omat<-read.table(paste0(folder,"roiOccipital_Lobesubfmri_",s, "_TRImatrix.txt"))
    
    subdata[ii,]<-c(i,Fsub$V1[1],sd(Fsub$V1[2:dim(Fsub)[1]]),'Frontal',length(Fsub$V1),mean(as.numeric(temp_dataset)),quantile(temp_dataset,.95),quantile(temp_dataset,.05),Fmat[1,1],Fmat[2,2],Fmat[2,1],Fmat[1,2],pvalue_perm=sum(Fsubsnull[i,]>Fsub$V1[1])/1000,ruletype[i])
    ii<-ii+1
    subdata[ii,]<-c(i,Osub$V1[1],sd(Osub$V1[2:dim(Osub)[1]]),'Occipital',length(Osub$V1),mean(as.numeric(temp_dataset2)),quantile(temp_dataset2,.95),quantile(temp_dataset2,.05),Omat[1,1],Omat[2,2],Omat[2,1],Omat[1,2],pvalue_perm=sum(Osubsnull[i,]>Osub$V1[1])/1000,ruletype[i])
    i = i+1
    ii<-ii+1   
}

subdata$roi<-as.factor(subdata$roi)
subdata$sid<-as.factor(subdata$sid)
subdata$ruletype<-as.factor(subdata$ruletype)
for (i in c(5:(length(subdata)-1))){
  subdata[,i] = as.numeric(subdata[,i])
}
for (i in c(2:3)){
  subdata[,i] = as.numeric(subdata[,i])
}
subdata$roi<-as.factor(subdata$roi)

# Add in some extra columns. 
for (i in c(1:nrow(subdata))){
  subdata$sig_perm[i] <- which.max(c(subdata$perm95[i],subdata$cvm[i]))-1
  b <- binom.test(subdata$cor1[i]+subdata$cor2[i],subdata$cor1[i]+subdata$cor2[i]+subdata$err1[i]+subdata$err2[i],0.5,"greater")
  subdata$pvalue_binom[i] = b$p.value
  subdata$sig_binom[i] <- which.max(c(subdata$pvalue_binom[i],0.05))-1
}
return(subdata)
}
```

# Plots and Analyses

#### What we were classifying

The information we were decoding in the frontal cortex was which rule participants used to respond to stimuli. For a comparison, we decoded visual features of the stimuli in the visual cortex. Conveniently, the rule’s in this task correlated with changes in visual features, so the partition of trials into classes was identical for visual and frontal lobe classification. For both regions we started with large ROI's to allow for the possibility that information was widely distributed, but we varied the size of the ROI as one of our parameters. 


## Standard Methods Choices

First, we attempted to decode information about rule in a large frontal cortex ROI using a standard set of analysis choices. Each voxel of unsmoothed fMRI data was detrended and z-scored within each run (leaving all time points). The patterns were then constructed using the 4th TR after the stimulus onset (rounded to nearest second). Classification analyses proceeded by treating each trial (single stimulus presentation) as a separate observation {y,x}. Used an Anova as univariate feature selector, thresholding at 500 voxels. The level of decoding was assessed using cross validation accuracy, using each run as a fold. 

Below, the average cross validation accuracy is shown for each subject. Error bars are 95 % confidence intervals centered around the mean an empirical null distribution generated for each subject via permutation, which we only calculated for a few of the parameter choice. 
(do these parameter choices coincide with the most popular ones from the meta-analysis?)

## Trying Different Methods

### Onset Time 

Our first consideration was whether decodability differed based on the time from stimulus onset. The hemodynamic response function varies in onset and shape in different regions, though the typically chosen 2-4 TR's is likely to reflect the peak in frontal cortex. Furthermore, its not clear whether decodability in the frontal cortex necessarily tracks the hrf, or whether it follows a different time course. 

Below we plot the cross validation for each subject using again a single TR, from TR's 0-8. (0-16seconds after stimulus onset). The dark red line is the average across subjects. 


* The peaks of classification are quite variable across subjects. I wonder if this aligns with the subject's HRF? If it does, it would be perhaps a good practice to estimate that first and use it choose the optimal time point. 

* the peak is much later than expected, however, this paper found that classification peak can occur later than raw signal peak. 

* To get even higher accuracy below, should have used the 5th TR instead of 4th. 

> In primary visual cortex (V1) stimulus information can be detected in the pattern of voxel responses more than a second before the average hemodynamic response of V1 deviates from baseline, and classification accuracy peaks before the peak of the average hemodynamic response. Both of these effects are restricted to early visual cortex, with higher level areas showing no difference or, in some cases, the opposite temporal relationship. (CITE)

### Classifier Choice

Each classifier differs in terms of its decision boundary shape and how it is estimated. Although the behavior of classifiers is well described theoretically, it is an empirical question of whether your data responds better to certain types of classifiers. ("no free lunch""). (see research notes)

* rdnf - random forest tree classifier
* linsvm - linear svm (C=1)
* knn - k-nearest neighbors
* plr - penalized logistic regression
* smlr - spare-multinomial logistic regression
* rbfsvm - nonlinear svm using radial basis function kernel
* gnb - guassian naive Bayes 
* fittedsvm - linear svm with C fit using additional cross validation

The results are expected. A linear svm with default parameter does well, along with GNB. Penalized logistic regression does do well. Nonlinear svm did not do any better, and actually seemed to do a little bit worse (similar to Misaka 2010). 
LDA does very poorly, but maybe be a fitting problem? See (2009 Pereira) for other comparison's of classifiers. 

For many of the classifiers we can also use implicit feature selection, regularization (shrinkage). For some, like logistic regression it may be necessary to fit with regularization. (hence bad performance here). I used another level of cross-validation (3 loops) to fit the amount of regularization in a linear SVM (the C parameter). This classifier did not did about the same in the visual cortex, and worse in the Frontal. Although I was concerned about this, Hastie finds the same thing.

> Without regularization it finds hyper plane with largest margin. Surprisingly, when the number of dimensions far exceeds the number of training examples the un-regularized often works about as well as best regularized version. Error rates are often insensitive to choice of c. (Hastie pg?).

### Does Movement matter? 

* Movement as measured by SPM realignment did not correlate with subject's classification accuracy. (Though that doesn't rule this out). 
* Also add movement regressor to GLM; that didn't help.

> motion effects, if motion is correlated with the design, might introduce major confounds into the signal. With multivariate analysis the problem becomes even more sever due to the high sensitivity of multivariate methods and the fact that motion effects might be impossible to eliminate entirely since they are strongly non-linear. So, even if you regress out whatever number of descriptors describing motion (mean displacement, angles, shifts, etc.) you would not be able to eliminate motion effects entirely. And that residual variance from motion spread through the entire volume might contribute to your generalization performance.



### Does RT confound our results?

Average different in RT between rules A and B did not correlate with a subjects classification accuracy. When regressed out like Todd suggested, results do decrease.
