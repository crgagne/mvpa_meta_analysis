```python
## Define the function 
def search_pubmed(foldername,enddate,search):
    import urllib, urllib2, sys
    import xml.etree.ElementTree as ET
    import unicodedata
    import os



    # some searches. 
    # MVPA + Frontal Lobe || MVPA + PFC || Decoding + Frontal Lobe || Classification  ... || Pattern ...
    #Affiliation [AD]
    #All Fields [ALL]
    #Author Name [AU]
    #Entrez Date [EDAT]
    #Issue [IP]
    #Journal Title [TA]
    #Language [LA]
    #MeSH Major Topic [MAJR]
    #MeSH Terms [MH]
    #Page Number [PG]
    #Personal Name as Subject [PS]
    #Publication Date [DP]
    #Publication Type [PT]
    #Subheadings [SH]
    #Text Words [TW]
    #Title/Abstract Words [TIAB]
    #Unique Identifiers [UID]

    def chunker(seq, size):
        return (seq[pos:pos + size] for pos in xrange(0, len(seq), size))



    ######## NEW SEARCH #########

    longquery = search

    ## Mega Search ##

    querytosave = 'MVPA_Search3'
    searchdir =  os.getcwd()
    searchdir = os.path.abspath(os.path.join(searchdir,'Data/'+foldername))
    esearch = 'http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&mindate=2001&maxdate='+enddate+'&retmode=xml&retmax=10000000&term=%s' % (longquery)
    handle = urllib.urlopen(esearch)
    data = handle.read()
    root = ET.fromstring(data)
    ids = [x.text for x in root.findall("IdList/Id")]
    print 'Got %d articles' % (len(ids))

    ## Save extract the titles, and save 
    if not os.path.exists(searchdir):
            os.makedirs(searchdir)
    text_file_titles = open(searchdir+'/q'+querytosave+'titles.txt',"w")  ## need to remove period***
    text_file_abstracts = open(searchdir+'/q'+querytosave+'abstracts.txt',"w")  ## need to remove period***

    for group in chunker(ids, 100):
        efetch = "http://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?&db=pubmed&retmode=xml&id=%s" % (','.join(group))
        handle = urllib.urlopen(efetch)
        data = handle.read()
        #print data
        root = ET.fromstring(data)
        for article in root.findall("PubmedArticle"): # finds the <PubmedArticles Tags>
            pmid = article.find("MedlineCitation/PMID").text
            year = article.find("MedlineCitation/Article/Journal/JournalIssue/PubDate/Year")
            title = article.find("MedlineCitation/Article/ArticleTitle").text
            authors = article.find("MedlineCitation/Article/AuthorList")
            stringToSave = ""
            stringToSaveA=""
            try:
                for author in authors:
                    lastname = author[0].text; 
                    try:
                        firstname = author[1].text; 
                    except:
                        firstname = ''; 
                    stringToSave=stringToSave+lastname+','+firstname+';'
            except:
                pass

            try:
                stringToSave = stringToSave+'('+year.text+')'
            except:
                pass

            stringToSave+=title
            stringToSave+='\n'


            print title
            try:
                abstract = article.find("MedlineCitation/Article/Abstract/AbstractText").text
             #   print abstract
            except:
                abstract = 'NA'


            ## deal with anoying unicode characters that cant be converted ascii
            if type(abstract) is unicode:
                abstract =unicodedata.normalize('NFKD', abstract).encode('ascii','ignore')

            stringToSaveA+=abstract
            stringToSaveA+='\n'


            try:
                stringToSave = unicodedata.normalize('NFKD', stringToSave).encode('ascii','ignore')
            except:
                pass 
            #if year is None: year = 'NA'
            #else: year = year.text
            #aulist = article.findall("MedlineCitation/Article/AuthorList/Author")
            #print pmid, year, len(aulist)

            # Write Abstracts Out #

            text_file_titles.write(stringToSave)
            text_file_abstracts.write(stringToSaveA)

    text_file_titles.close()
    text_file_abstracts.close()  
```

```python
## Conduct the search ##
search = '(whole brain [TW] OR entire brain [TW] OR prefrontal [TW] OR frontal [TW] OR PFC [TW] OR frontoparietal [TW]) AND (functional magnetic resonance [TW] OR fMRI [TW] OR functional imaging [TW] OR neuroimaging [TW] OR magnetic resonance imaging [TW]) AND (decoded [TW] OR decoding [TW] OR decodes [TW]  OR multivariate analyses [TW] OR multivoxel pattern analysis [TW] OR multi-voxel pattern analysis [TW] OR MVPA [TW] OR multivariate pattern analysis [TW] OR multivariate pattern classification [TW] OR Classification [TW] OR Classifying [TW] OR Representational Similarity Analysis [TW] OR RSA [TW]) NOT (children [TW] OR attention-deficit [TW] OR Epilepsy [TW] OR disease [TW] OR autism [TW] OR aging [TW] OR depression [TW] OR anxiety [TW] OR clinical [TW] OR schizophrenia [TW] OR patients [TW] OR syndrome [TW] OR aged [TW] OR dyslexia [TW] OR dysplasia [TW] OR tumor [TW])' 
enddate = '2014'
foldername = 'Data_Searchs_Jan'
search_pubmed(foldername,enddate,search)
```
