# Load Back In 

```python
boot_results_vc = np.load('../data_meta_analysis/data_derived_meta_analysis/bootstrap_results_visual_cortex.npz')
boot_results_vt = np.load('../data_meta_analysis/data_derived_meta_analysis/bootstrap_results_ventral_temporal.npz')
boot_results_both = np.load('../data_meta_analysis/data_derived_meta_analysis/bootstrap_results_both_visual.npz')
```

```python
x = np.linspace(0,1,100) # what range for cdf # 
epcdf_store_vt = boot_results_vt['epcdf_store']
eppdf_store_vt = boot_results_vt['eppdf_store']
meann_vt = boot_results_vt['meann']
mediann_vt = boot_results_vt['mediann']
meann_study_vt=boot_results_vt['meann_study']


epcdf_store_vc = boot_results_vc['epcdf_store']
eppdf_store_vc = boot_results_vc['eppdf_store']
meann_vc = boot_results_vc['meann']
mediann_vc = boot_results_vc['mediann']
meann_study_vc=boot_results_vc['meann_study']

epcdf_store_both = boot_results_both['epcdf_store']
eppdf_store_both = boot_results_both['eppdf_store']
meann_both = boot_results_both['meann']
mediann_both = boot_results_both['mediann']
meann_study_both=boot_results_both['meann_study']



```

# CDF and PDF Plots 

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize = (12,4))
ax1.step(x, epcdf_store_vt.T);
ax2.plot(x, eppdf_store_vt.T);


fig, (ax1,ax2) = plt.subplots(1, 2, figsize = (12,4))
ax2.plot(x, eppdf_store_vc.T);
ax1.step(x, epcdf_store_vc.T);

fig, (ax1,ax2) = plt.subplots(1, 2, figsize = (12,4))
ax2.plot(x, eppdf_store_both.T);
ax1.step(x, epcdf_store_both.T);
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize = (12,4))

upci = np.percentile(epcdf_store_vc,97.5,axis=0)
loci = np.percentile(epcdf_store_vc,2.5,axis=0)
ax1.plot(x,np.mean(epcdf_store_vc,axis=0),linewidth=2,color='black')
ax1.fill_between(x,upci,loci)

upci = np.percentile(epcdf_store_vt,97.5,axis=0)
loci = np.percentile(epcdf_store_vt,2.5,axis=0)
ax1.plot(x,np.mean(epcdf_store_vt,axis=0),linewidth=2,color='black')
ax1.fill_between(x,upci,loci,color='green')

##### PDF 

upci = np.percentile(eppdf_store_vc,97.5,axis=0)
loci = np.percentile(eppdf_store_vc,2.5,axis=0)
ax2.plot(x,np.mean(eppdf_store_vc,axis=0),linewidth=2,color='black')
ax2.fill_between(x,upci,loci,label='visual cortex')

upci = np.percentile(eppdf_store_vt,97.5,axis=0)
loci = np.percentile(eppdf_store_vt,2.5,axis=0)
ax2.plot(x,np.mean(eppdf_store_vt,axis=0),linewidth=2,color='black')
ax2.fill_between(x,upci,loci,color='green',label='ventral temporal')

ax1.set_title('Cumulative Dist. Func.')
ax2.set_title('Prob. Dens. Func.')

ax1.set_xlabel('Accuracy')
ax2.set_xlabel('Accuracy')
plt.legend()
plt.suptitle('Distribution of Accuracies-Boot-strapped CI',fontsize=12)
#plt.tight_layout()
#plt.savefig('distributions_cdf_pdf_boot_visual.png')
```
