```python
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings("ignore")

sns.set_style('white')
# params = {
#    'axes.labelsize': 10,
#    'axes.titlesize': 10,
#     'axes.titleweight': 'normal',
#     'axes.labelweight':'normal',
#    'text.fontsize': 10,
#    'legend.fontsize': 10,
#    'xtick.labelsize': 10,
#    'ytick.labelsize': 10,
#    'text.usetex': False,
# }

params = {
   'axes.labelsize': 8,
   'axes.titlesize': 8,
    'axes.titleweight': 'normal',
    'axes.labelweight':'normal',
   'text.fontsize': 8,
   'legend.fontsize': 7,
   'xtick.labelsize': 8,
   'ytick.labelsize': 8,
   'text.usetex': False,
    'text.color':'black'
}

plt.rcParams.update(params)
plt.rcParams['font.sans-serif']='Helvetica'
plt.rcParams['font.family']='sans-serif'

fig = plt.figure(figsize=(5,3.3))
alphadots=1.0
sizedots=6
mainlinewidth = 1
whiteoutline = 0.1
median_line_widths=1
chanceline = 1
alphaa =0.7


gs = gridspec.GridSpec(10, 8)
ax1 = plt.subplot(gs[0:5,0:4])
ax2 = plt.subplot(gs[0:5,4:8])
ax3 = plt.subplot(gs[5:10,0:4])
ax4 = plt.subplot(gs[5:10,4:8])


#### CDF 


# frontal all 
upci = np.percentile(epcdf_store_frontal_all[40:100],97.5,axis=0)
loci = np.percentile(epcdf_store_frontal_all[40:100],2.5,axis=0)
ax1.plot(x[40:100],np.mean(epcdf_store_frontal_all,axis=0)[40:100],linewidth=mainlinewidth,color=frontalb)
ax1.fill_between(x[40:100],upci[40:100],loci[40:100],color=frontalb,alpha=alphaa,linestyle='None',linewidth=0)

# frontal rule 
upci = np.percentile(epcdf_store_frontal_rule[40:100],97.5,axis=0)
loci = np.percentile(epcdf_store_frontal_rule[40:100],2.5,axis=0)
ax3.plot(x[40:100],np.mean(epcdf_store_frontal_rule,axis=0)[40:100],linewidth=mainlinewidth,color=frontalrule)
ax3.fill_between(x[40:100],upci[40:100],loci[40:100],color=frontalrule,alpha=alphaa,linestyle='None',linewidth=0)
#ax3.plot(x[40:100],upci[40:100],color='white',linewidth=whiteoutline)
#ax3.plot(x[40:100],loci[40:100],color='white',linewidth=whiteoutline)


# visual cortex 
upci = np.percentile(epcdf_store_vc[40:100],97.5,axis=0)
loci = np.percentile(epcdf_store_vc[40:100],2.5,axis=0)
ax1.plot(x[40:100],np.mean(epcdf_store_vc,axis=0)[40:100],linewidth=mainlinewidth,color=visualc)
ax1.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualc,alpha=alphaa,linestyle='None',linewidth=0)
ax3.plot(x[40:100],np.mean(epcdf_store_vc,axis=0)[40:100],linewidth=mainlinewidth,color=visualc)
ax3.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualc,alpha=alphaa,linestyle='None',linewidth=0)

#  ventral temporal 
upci = np.percentile(epcdf_store_vt[40:100],97.5,axis=0)
loci = np.percentile(epcdf_store_vt[40:100],2.5,axis=0)
ax1.plot(x[40:100],np.mean(epcdf_store_vt,axis=0)[40:100],linewidth=mainlinewidth,color=visualt)
ax1.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualt,alpha=alphaa,linestyle='None',linewidth=0)
ax3.plot(x[40:100],np.mean(epcdf_store_vt,axis=0)[40:100],linewidth=mainlinewidth,color=visualt)
ax3.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualt,alpha=alphaa,linestyle='None',linewidth=0)


############
##### PDF 

upci = np.percentile(eppdf_store_frontal_all[40:100],97.5,axis=0)
loci = np.percentile(eppdf_store_frontal_all[40:100],2.5,axis=0)
ax2.plot(x[40:100],np.mean(eppdf_store_frontal_all,axis=0)[40:100],linewidth=mainlinewidth,color=frontalb)
ax2.fill_between(x[40:100],upci[40:100],loci[40:100],label='Frontal (all)',color=frontalb,alpha=alphaa,linestyle='None',linewidth=0)
ax2.fill_between(x[50:52],[2,2],[2.001,2.001],color=frontalrule,
                 alpha=alphaa,label='Frontal (rule)',linestyle='None',linewidth=0)

upci = np.percentile(eppdf_store_frontal_rule[40:100],97.5,axis=0)
loci = np.percentile(eppdf_store_frontal_rule[40:100],2.5,axis=0)
ax4.plot(x[40:100],np.mean(eppdf_store_frontal_rule,axis=0)[40:100],linewidth=mainlinewidth,color=frontalrule)
ax4.fill_between(x[40:100],upci[40:100],loci[40:100],color=frontalrule,alpha=alphaa,linestyle='None',linewidth=0)


upci = np.percentile(eppdf_store_vc[40:100],97.5,axis=0)
loci = np.percentile(eppdf_store_vc[40:100],2.5,axis=0)
ax2.plot(x[40:100],np.mean(eppdf_store_vc,axis=0)[40:100],linewidth=mainlinewidth,color=visualc)
ax2.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualc,label='Occipital',alpha=alphaa,linestyle='None',linewidth=0)
ax4.plot(x[40:100],np.mean(eppdf_store_vc,axis=0)[40:100],linewidth=mainlinewidth,color=visualc)
ax4.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualc,alpha=alphaa,linestyle='None',linewidth=0)

upci = np.percentile(eppdf_store_vt[40:100],97.5,axis=0)
loci = np.percentile(eppdf_store_vt[40:100],2.5,axis=0)
ax2.plot(x[40:100],np.mean(eppdf_store_vt,axis=0)[40:100],linewidth=mainlinewidth,color=visualt)
ax2.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualt,label='Ventral temporal',alpha=alphaa,linestyle='None',linewidth=0)
ax4.plot(x[40:100],np.mean(eppdf_store_vt,axis=0)[40:100],linewidth=mainlinewidth,color=visualt)
ax4.fill_between(x[40:100],upci[40:100],loci[40:100],color=visualt,alpha=alphaa,linestyle='None',linewidth=0)




#ax1.axvline(x=0.5,color='k',linestyle='--',linewidth=chanceline)
ax2.axvline(x=0.5,color='k',linestyle='--',linewidth=chanceline)
#ax3.axvline(x=0.5,color='k',linestyle='--',linewidth=chanceline)
ax4.axvline(x=0.5,color='k',linestyle='--',label='Chance decoding',linewidth=chanceline)

ax2.axvline(x=mediann_frontal_all.mean(),color=frontalb,linewidth=median_line_widths)
ax2.axvline(x=mediann_vt.mean(),color=visualt,linewidth=median_line_widths)
ax2.axvline(x=mediann_vc.mean(),color=visualc,linewidth=median_line_widths)
plt.sca(ax2)
plt.legend(bbox_to_anchor=(.55, 1.), loc=2)

ax4.axvline(x=mediann_frontal_rule.mean(),color=frontalrule,linewidth=median_line_widths)
ax4.axvline(x=mediann_vt.mean(),color=visualt,linewidth=median_line_widths)
ax4.axvline(x=mediann_vc.mean(),color=visualc,linewidth=median_line_widths)


#ax1.set_xlabel('Decoding Accuracy')
ax1.set_ylabel('Cumulative probability',fontname='Helvetica')
ax3.set_xlabel('Decoding accuracy',fontname='Helvetica')
ax3.set_ylabel('Cumulative probability',fontname='Helvetica')

#ax2.set_xlabel('Decoding Accuracy')
ax2.set_ylabel('Probability density',fontname='Helvetica')
ax4.set_xlabel('Decoding accuracy',fontname='Helvetica')
ax4.set_ylabel('Probability density',fontname='Helvetica')

# ax1.set_title('a',fontname='Helvetica')
# ax2.set_title('b',fontname='Helvetica')
# ax3.set_title('c',fontname='Helvetica')
# ax4.set_title('d',fontname='Helvetica')

ax2.set_yticks([])
ax4.set_yticks([])
ax2.set_yticklabels([])
ax4.set_yticklabels([])

for ax in [ax1,ax2,ax3,ax4]:
    ax.spines["top"].set_visible(False)  
    ax.spines["right"].set_visible(False) 
    ax.set_axis_bgcolor('white')


plt.tight_layout(h_pad=2.0,w_pad=1.0)

# Save the image in memory in PNG format
png1 = BytesIO()
fig.savefig(png1, format='png',dpi=1000)

# Load this image into PIL
png2 = Image.open(png1)

# (3) save as TIFF
png2.save('../Figures_JOCN_Final/Figure2_v8b.tiff')
png1.close()

#plt.savefig('../Figures/Figure1_v8b.png',dpi=400)
#plt.savefig('../Figures/Figure1_v8b.eps',dpi=400)
#plt.show()
```
