```r
coefs1.C<-c()
coefs2.C<-c()
coefs3.C<-c()
inter.C<-c()
se3.C<-c()

coefs12.C<-c()

for (i in seq(200)){
r1 = rnorm(.5,.05,n=10)
r2 = rnorm(.55,.05,n=10)
r3 = rnorm(.65,.05,n=10)
y = c(r1,r2,r3)

## know where observations came from 
x1 = c(rep(1,10),rep(0,10),rep(0,10))
x2 = c(rep(0,10),rep(1,10),rep(0,10))
x3 = c(rep(0,10),rep(0,10),rep(1,10))
x12 = as.integer(c(x1 | x2))
x1[8:12]=1
x2[8:12]=1
s1<-summary(lm(y~x12))
s1<-summary(lm(y~x1+x2+x3))
#s1
#mean(y)

coefs1.C<-cbind(coefs1.C,s1$coefficients[1,1]+s1$coefficients[2,1])
inter.C<-cbind(inter.C,s1$coefficients[1,1])
coefs2.C<-cbind(coefs2.C,s1$coefficients[1,1]+s1$coefficients[3,1])
coefs3.C<-cbind(coefs3.C,s1$coefficients[1,1]+s1$coefficients[4,1])
#coefs12.C<-cbind(coefs12.C,s1$coefficients[1,1]+s1$coefficients[2,1]+s1$coefficients[3,1])
    
#se3.C<-cbind(se3.C,s1$coefficients[3,2]) 
#se3.C<-cbind(se3.C,sqrt(s1$coefficients[1,2]**2+s1$coefficients[4,2]**2))    

}
```
