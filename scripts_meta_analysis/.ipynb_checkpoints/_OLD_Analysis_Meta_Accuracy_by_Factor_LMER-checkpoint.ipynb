```r
datadir = '../Data'
data = read.csv(paste(datadir,"Data_Classifications_v3.csv",sep='/'))
datapapers = read.csv(paste(datadir,"Data_Papers_v2.csv",sep='/'))

drops<-names(data)[grep("AAL",names(data))]
data<-data[,!(names(data) %in% drops)]
drops<-names(data)[grep("Notes",names(data))]
data<-data[,!(names(data) %in% drops)]
drops<-names(data)[grep("Coordinate",names(data))]
data<-data[,!(names(data) %in% drops)]
#drops<-names(data)[grep("Region.Label",names(data))]
#data<-data[,!(names(data) %in% drops)]
idx = !is.na(data$PID)
data<-data[idx,]

data$PID = as.factor(data$PID) # make sure paper ID is factor not numeric

data$Classifier.type = relevel(data$Classifier.type,'GNB')


drops<-names(datapapers)[grep("Subcategory",names(datapapers))]
datapapers<-datapapers[,!(names(datapapers) %in% drops)]
drops<-names(datapapers)[grep("Abstract",names(datapapers))]
datapapers<-datapapers[,!(names(datapapers) %in% drops)]
drops<-names(datapapers)[grep("Keyword",names(datapapers))]
datapapers<-datapapers[,!(names(datapapers) %in% drops)]
redo=1


data<-data[!is.na(data$Accuracy.50),]

data<-data[data$Roi.Search=='ROI',]

data$PID<-as.factor(data$PID)
data$Info.type<-as.factor(data$Info.type)

```

```r
fit<-lm(Accuracy.50 ~ 1+Info.type, data=data) # this is correct code according to gelman. For random Intercept. 
summary(fit)
```

```r
fit1<-lmer(data$Accuracy.50 ~ data$Info.type + (1|data$PID)) # this is correct code according to gelman. For random Intercept. 
summary(fit1)
```
