## Load Data

```r
data = read.csv("../Data/Data_Classifications_v7.csv")

data<-data[c('PID','Info.type','Accuracy.50','Significance','Roi.Search',
             'Pattern.type','Classifier.type','Scanner','Smoothing','Coregistration','Pattern.normalization.type','Num_subs')]

factors<-c('Info.type','Significance','Roi.Search',
             'Pattern.type','Classifier.type','Scanner','Smoothing','Coregistration','Pattern.normalization.type','Num_subs')

for (pid in unique(data$PID)){
    #print('--')
    #print(pid)
    #print(nrow(data1[data1$PID==pid,]))
    #print('--')
    data[data$PID==pid,'n_per_study']=nrow(data[data$PID==pid,])
}


# msc

R1 = read.csv(paste("X_region1_before_clustering_2.csv"))
#R2 = read.csv(paste("X_region1_after_clustering.csv"))
R3 = read.csv(paste("X_region1_after_clustering_lr_2.csv"))
R4 = read.csv(paste("X_region1_after_clustering2_2.csv"))


R1names = colnames(R1)
#R2names = colnames(R2)
R3names = colnames(R3)
R4names = colnames(R4)

i=1
for (name in R3names){
    R3names[i]<-paste(sep='','R3',name)
    i=i+1
}
R3names
colnames(R3)<-R3names
head(R3)

i=1
for (name in R4names){
    R4names[i]<-paste(sep='','R4',name)
    i=i+1
}
R4names
colnames(R4)<-R4names
head(R4)
```

```r
unique(data$Pattern.normalization.type)
```

```r
data = cbind(data,R1,R3,R4) # will bind with other region names too 


# mscl fixed 
data<-data[!data$Classifier.type=='neural-net',] # no observations but still has a level
data[data$Scanner=='4T','Scanner']='3T' # only one observatoin with 3T

data$PID<-as.factor(data$PID)
data$Info.type<-as.factor(data$Info.type)
data$Smoothing<-as.factor(data$Smoothing)
data$Roi.Search<-as.factor(data$Roi.Search)
data$Classifier.type<-as.factor(data$Classifier.type)
data$Coregistration<-as.factor(data$Coregistration)
data$Pattern.type<-as.factor(data$Pattern.type)

#data[data$Pattern.normalization.type=='Normalized Spatially']

data$Pattern.normalization.type<-as.factor(data$Pattern.normalization.type)

data<-droplevels(data)

data_50<-subset(data,!is.na(Accuracy.50))

data_50_non<-subset(data,!is.na(Accuracy.50)& Significance==0)

data_50_sig<-subset(data,!is.na(Accuracy.50) & Significance==1)
head(data_50_sig)
nrow(data_50_sig)


# make unique
keys <- colnames(data_50_sig)[!grepl('Accuracy.50',colnames(data_50_sig))] # grab everything but accuracy
X <- as.data.table(data_50_sig)
data_50_sig_unique<- X[,list(Accuracy.50.mean= mean(Accuracy.50)),keys] # average the non-unique
data_50_sig_unique$Accuracy.50<-data_50_sig_unique$Accuracy.50.mean
data_50_sig_unique<-data.frame(data_50_sig_unique)

keys <- colnames(data_50)[!grepl('Accuracy.50',colnames(data_50))] # grab everything but accuracy
X <- as.data.table(data_50)
data_50_unique<- X[,list(Accuracy.50.mean= mean(Accuracy.50)),keys] # average the non-unique
data_50_unique$Accuracy.50<-data_50_unique$Accuracy.50.mean
data_50_unique<-data.frame(data_50_unique)



nrow(data_50_sig_unique)
nrow(data_50_unique)
```

```r
d<-data_50_sig[c('Accuracy.50','PID',R1names)]
head(d)
nrow(d)
```

```r
# Plot without CI 
X<-as.matrix(data_50_sig[(R1names)])
y<-as.matrix(data_50_sig['Accuracy.50'])


mean_acc<-t(X)%*%y/colSums(X) # average the accuracy per region.

mean_acc<-data.frame(mean_acc) 
mean_acc$region<-rownames(mean_acc) # make column name the row name
head(mean_acc)

options(repr.plot.width=8, repr.plot.height=3)
ggplot(data=mean_acc,aes(x=region,y=Accuracy.50))+geom_point()+theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
n_boots = 500
mean_acc<-t(X)%*%y/colSums(X) # average the accuracy per region.
mean_acc_all_boots<-mean_acc # start with the real data 

for (boot in seq(n_boots)){
    
# get a new copy of d each time 
d<-data_50_sig[c('Accuracy.50','PID',R1names)]
PIDs<-unique(d$PID) 

# sample new studies with replacement
PIDs_boot<-sample(PIDs,size=length(PIDs),replace = TRUE)

# loop trhough the newly sampled set of studies 
for (pid in PIDs_boot){
study_obs<-d[d$PID==pid,] # get the rows for that study 
study_n<-nrow(study_obs) # get the number 
study_obs_rows_boot<-sample(seq(study_n),size=study_n,replace=TRUE) # sample from those 
study_obs_boot<-study_obs[study_obs_rows_boot,]
d[d$PID==pid,]<-study_obs_boot # put back into dataset  
}

Xb<-as.matrix(d[(R1names)])
yb<-as.matrix(d['Accuracy.50'])

mean_acc<-t(Xb)%*%yb/colSums(Xb) # average the accuracy per region.

mean_acc_all_boots<-cbind(mean_acc_all_boots,mean_acc)
}


```

```r
head(mean_acc_all_boots)
```

```r
X<-as.matrix(data_50_sig[(R1names)])
y<-as.matrix(data_50_sig['Accuracy.50'])

# calcualte actual
mean_acc<-t(X)%*%y/colSums(X) # average the accuracy per region.
mean_acc<-data.frame(mean_acc) 

# get the boostrapped SD 
mean_acc_all_boots_df<-data.frame((mean_acc_all_boots))
boots_sd<-apply(mean_acc_all_boots_df, 1, sd)
mean_acc<-data.frame(cbind(mean_acc,boots_sd))

# add row name
mean_acc$region<-rownames(mean_acc) # make column name the row name


head(mean_acc)

# Plot
options(repr.plot.width=8, repr.plot.height=5)
ggplot(data=mean_acc,aes(x=region,y=Accuracy.50))+geom_point()+theme(axis.text.x = element_text(angle = 45, hjust = 1))+geom_errorbar(aes(ymin=Accuracy.50-boots_sd, ymax=Accuracy.50+boots_sd))
```
