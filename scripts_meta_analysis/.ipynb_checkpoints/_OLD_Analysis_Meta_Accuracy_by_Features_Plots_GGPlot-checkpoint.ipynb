### Distribution of classification accuracy by study

Rather than assume each classification attempt within a study is independent we can look at the average accuracy for each study..

```python
%%R 
databystudy <- aggregate( Accuracy.50~PID, data, mean ) ### THIS WILL ONLY SHOW 50% studies so there will be less. 
p <- ggplot(data=databystudy, aes(x=Accuracy.50,fill=1)) +geom_histogram(alpha=0.7)+theme(legend.position="bottom")+geom_vline(xintercept = mean(databystudy$Accuracy.50[!is.na(databystudy$Accuracy.50)]))
p
```

### What are the studies that yield the highest classifications? 

First, its interesting just to look at the top observations (the paper they came from, and the type of information they were decoding). Notably they are from only a few studies. And, at least some of the top classifications seem to distinguish fundamentally different cognitive processes. For example paper 48 classified whether participants were using saccadding or reaching. They decoded during planning and execution phase of the trial, and the execution phase results are what are here in the top classifications. Although classification success could reflect the the same neurons coding for these two tasks, it is more likely that the high success is due to different populations of neurons responding for each of these two tasks.

## Proper Statistical Models? 

Since each study contributes several observations of classification accuracy, the data is considered hierarchical. Therefore, I am going to use a linear mixed model, treating study as a random effect. This assumes there is something related among observations from the same study, other than the variables I recorded, that contributes to classification accuracy (i.e. data quality). In the model this assumed relationship amounts to adding an extra intercept is per study, each of which is drawn from the same zero-centered Gaussian distribution.
