# Plot And Boot-strap

```r
# Plot without CI 
X<-as.matrix(d[(R1names)])
y<-as.matrix(d['Accuracy.50'])


mean_acc<-t(X)%*%y/colSums(X) # average the accuracy per region.

mean_acc<-data.frame(mean_acc) 
mean_acc$region<-rownames(mean_acc) # make column name the row name
head(mean_acc)

jpeg('test.jpg',width=10.25,height=6,units="in",res=800)
options(repr.plot.width=16, repr.plot.height=7)
ggplot(data=mean_acc,aes(x=region,y=Accuracy.50))+geom_point()+theme(axis.text.x = element_text(angle = 90, hjust = 1))
dev.off()
ggplot(data=mean_acc,aes(x=region,y=Accuracy.50))+geom_point()+theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

```r
n_boots = 100
mean_acc<-t(X)%*%y/colSums(X) # average the accuracy per region.
mean_acc_all_boots<-mean_acc # start with the real data 

for (boot in seq(n_boots)){
    
# get a new copy of d each time 
d<-d[c('Accuracy.50','PID',R1names)]
PIDs<-unique(d$PID) 

# sample new studies with replacement
PIDs_boot<-sample(PIDs,size=length(PIDs),replace = TRUE)

# loop trhough the newly sampled set of studies 
for (pid in PIDs_boot){
study_obs<-d[d$PID==pid,] # get the rows for that study 
study_n<-nrow(study_obs) # get the number 
study_obs_rows_boot<-sample(seq(study_n),size=study_n,replace=TRUE) # sample from those 
study_obs_boot<-study_obs[study_obs_rows_boot,]
d[d$PID==pid,]<-study_obs_boot # put back into dataset  
}

Xb<-as.matrix(d[(R1names)])
yb<-as.matrix(d['Accuracy.50'])

mean_acc<-t(Xb)%*%yb/colSums(Xb) # average the accuracy per region.

mean_acc_all_boots<-cbind(mean_acc_all_boots,mean_acc)
}


```

```r
head(mean_acc_all_boots)
```

```r
X<-as.matrix(d[(R1names)])
y<-as.matrix(d['Accuracy.50'])

# calcualte actual
mean_acc<-t(X)%*%y/colSums(X) # average the accuracy per region.
mean_acc<-data.frame(mean_acc) 

# get the boostrapped SD 
mean_acc_all_boots_df<-data.frame((mean_acc_all_boots))
boots_sd<-apply(mean_acc_all_boots_df, 1, sd)
mean_acc<-data.frame(cbind(mean_acc,boots_sd))

# add row name
mean_acc$region<-rownames(mean_acc) # make column name the row name


head(mean_acc)

# Plot
options(repr.plot.width=8, repr.plot.height=5)
ggplot(data=mean_acc,aes(x=region,y=Accuracy.50))+geom_point()+theme(axis.text.x = element_text(angle = 90, hjust = 1))+geom_errorbar(aes(ymin=Accuracy.50-boots_sd, ymax=Accuracy.50+boots_sd))
```

```r
d<-data_50_sig[c('Accuracy.50','PID',R1names)]
d=d[rowSums(d[,R1names])==1,]
nrow(d)

gmean<-mean(d$Accuracy.50)
print(gmean)
d$Accuracy.50_wo_gmean<-d$Accuracy.50-gmean
region_model<-lm(as.formula(paste('Accuracy.50_wo_gmean~-1+',R1names_string,sep='')),data=d)
summary(region_model)
```

```r
options(repr.plot.width=8, repr.plot.height=4)
fits<-c()
se<-c()
i=1
for (name in R1names){
    ef<-effect(name,region_model)
    fits[i]<-tail(ef$fit, n=1)
    se[i]<-tail(ef$se, n=1)
    i=i+1
}
dd = data.table(name=R1names,se=se,Accuracy=fits)
ggplot(aes(y = Accuracy, x = name,ymax = Accuracy + se, ymin=Accuracy - se), data = dd) +geom_point()+geom_errorbar()+theme(axis.text.x = element_text(angle = 45, hjust = 1))+ggtitle('Region Only Model')

```
