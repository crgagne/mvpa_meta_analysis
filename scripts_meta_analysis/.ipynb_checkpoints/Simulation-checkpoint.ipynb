```r
library(ggplot2)
library(nlme)
```

```r
r1=3
r2=1

studies = c()
accs = c()
regions = c()
for (region in seq(2)){ 
    for (study in seq(10)){
        for (sample in seq(10)){
            studies<-c(studies,study)
            if(region==1){
                u=r1
            }else{
                u=r2
            }
            accs<-c(accs,rnorm(1,u,1))
            regions<-c(regions,region)
        }
    }
}
d<-data.frame(cbind(accs,studies,regions))
d$regions<-as.factor(d$regions)
d$studes<-as.factor(d$studies)
```

```r
ggplot(data=d,aes(x=studies,y=accs,color=regions))+geom_point()
```

```r
summary(lme(accs~regions,random=~1|studies,data=d))
```

```r
r1=3
r2=1

studies = c()
accs = c()
regions = c()
    for (study in seq(10)){
        smean<-rnorm(1,0,.5)
        for (sample in seq(10)){
            studies<-c(studies,study)
            if(study>5){
                u=r1
                region=1
            }else{
                region=2
                u=r2
            }
            accs<-c(accs,rnorm(1,u,1)+smean)
            regions<-c(regions,region)
        }
    }

d<-data.frame(cbind(accs,studies,regions))
d$regions<-as.factor(d$regions)
d$studes<-as.factor(d$studies)
d$accs<-d$accs-mean(d$accs)
```

```r
ggplot(data=d,aes(x=studies,y=accs,color=regions))+geom_point()
```

```r
mod<-lm(accs~-1+regions,data=d)
summary(mod)
```

```r
mod<-lme(accs~-1+regions,random=~1|studies,data=d,method=)
summary(mod)
```

```r

r1=3
r2=1

studies = c()
accs = c()
regions = c()
    for (study in seq(10)){
        smean<-rnorm(1,0,.5)
        for (sample in seq(10)){
            studies<-c(studies,study)
            if(study>5){
                u=r1
                region=1
            }else{
                region=2
                u=r2
            }
            accs<-c(accs,rnorm(1,u,1)+smean)
            regions<-c(regions,region)
        }
    }


for (region in seq(2)){ 
    for (study in seq(10)){
        smean<-rnorm(1,0,.5)
        for (sample in seq(10)){
            studies<-c(studies,study+10)
            if(region==1){
                u=r1
            }else{
                u=r2
            }
            accs<-c(accs,rnorm(1,u,1)+smean)
            regions<-c(regions,region)
        }
    }
}


d<-data.frame(cbind(accs,studies,regions))
d$regions<-as.factor(d$regions)
d$studes<-as.factor(d$studies)
d$accs<-d$accs-mean(d$accs)
ggplot(data=d,aes(x=studies,y=accs,color=regions))+geom_point()
ggplot(data=d,aes(color=studies,y=accs,x=regions))+geom_point()
```
