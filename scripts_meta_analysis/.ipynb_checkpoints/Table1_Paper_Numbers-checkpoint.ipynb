```r
data = read.csv("../data_meta_analysis/Data_Classifications_v8b.csv")
length(unique(data$PID))
```

```r
PID=12
```

```r
data<-data[data$Significance==1,] 
```

```r
for (PID in unique(data$PID)){
    data[data$PID==PID,'num_analyses']<-nrow(data[data$PID==PID,])
    data[data$PID==PID,'has_rule']<-is.element('Rule',data[data$PID==PID,'Info.type'])
    data[data$PID==PID,'study_acc']<-round(mean(data[data$PID==PID,'Accuracy.50']),3)
    
}
```

```r
d<-data[!duplicated(data$PID),c('PID','Title','Num_subs','num_analyses','has_rule','study_acc')]
d<-d[order(d$PID),]
#d[0:40,]
d[40:80,]
```
