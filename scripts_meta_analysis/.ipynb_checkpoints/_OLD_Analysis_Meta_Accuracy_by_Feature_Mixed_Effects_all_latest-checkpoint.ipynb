## Mixed Effects: Region +  All, Averaging within Region 


```r
## remove neural-net
data<-data[!data$Classifier.type=='neural-net',]
```

```r
data$Region<-as.factor(data$Region)
data$PID<-as.factor(data$PID)
data$Info.type<-as.factor(data$Info.type)
data$Smoothing<-as.factor(data$Smoothing)
data$Roi.Search<-as.factor(data$Roi.Search)
data$Classifier.type<-as.factor(data$Classifier.type)
data$Coregistration<-as.factor(data$Coregistration)
data$Pattern.type<-as.factor(data$Pattern.type)
data$Pattern.normalization.type<-as.factor(data$Pattern.normalization.type)
```

```r
data$Classifier.type<-relevel(data$Classifier.type,'correlation')
```

```r
data<-droplevels(data)
```

```r
levels(data$Region)
levels(data$PID)
levels(data$Info.type)
levels(data$Smoothing)
levels(data$Roi.Search)
levels(data$Classifier.type)
levels(data$Coregistration)
levels(data$Pattern.type)
levels(data$Pattern.normalization.type)
```

```r
variables = c('Accuracy.50','PID','Info.type','Smoothing','Roi.Search','Classifier.type','Coregistration','Pattern.type','Pattern.normalization.type','Region','Significance')
```

subset to 50% accuracy

```r
data_50<-data[!is.na(data$Accuracy.50),variables]
nrow(data_50)
data_50<-data_50[complete.cases(data_50),]
nrow(data_50)
```

```r
head(data_50)
nrow(data_50)
```

average just the repeats within region. Now there are only 181 rows. 

**may want to do differently here**

```r
keys <- colnames(data_50)[!grepl('Accuracy.50',colnames(data_50))]
keys
```

```r
X <- as.data.table(data_50)
```

```r
data_50_mean<- X[,list(Accuracy.50.mean= mean(Accuracy.50)),keys]
```

```r
data_50_mean<-data.frame(data_50_mean)
head(data_50_mean)
print(nrow(data_50_mean))
```

plot these results 

dummy variable coding. 

```r
model = lme(Accuracy.50.mean~Info.type+Region+Roi.Search+Classifier.type+Smoothing+Coregistration+Pattern.type+Pattern.normalization.type+Significance,random = ~1|PID,data=data_50_mean)
summary(model)
```

Wald F test instead. Tests sequentially, but that doesnt matter here. 

```r
anova(model)
```

plot coefficients

```r
s<-summary(model)
coef<-s$tTable[,1] 
se<-s$tTable[,2] 
df<-data.frame(cbind(coef,se))
#print(s$tTable[,0])
df$name<-rownames(df)
df<-df[2:nrow(df),]
```

```r
## save to csv
```

```r
write.csv(df,'feature_mixed_effects_all.csv')
```

```r

ggplot(df, aes(y=coef,x=name))+geom_point()+geom_errorbar(aes(ymax = coef + se, ymin=coef- se))+theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
library(lme4)
```

```r
model = lmer(Accuracy.50.mean~Info.type+Region+(1|PID),data=data_50_mean)
#summary(model)
```

```r
#install.packages('influence.ME', repos='http://cran.us.r-project.org')
```

```r

infl <- influence(model, obs = TRUE)
#cooks.distance(infl)
```

```r
plot(infl, which = "cook")
```

```r
X = model.matrix(model,data=data_50_mean)
X<-X[,2:ncol(X)]
C = cor(X)
```

```r
options(repr.plot.width=20, repr.plot.height=20)


col <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))
col <- colorRampPalette(c("#4477AA", "#77AADD", "#FFFFFF","#EE9988", "#BB4444"))

corrplot(C, method="color", col=col(200),  
         type="upper", order="hclust", 
         addCoef.col = "black", # Add coefficient of correlation
         tl.col="black", tl.srt=90, #Text label color and rotation
         # Combine with significance
         # hide correlation coefficient on the principal diagonal
         diag=FALSE 
         )
```

```r
options(repr.plot.width=8, repr.plot.height=8)

image(t(X))
```
