```r
calc_median<-function(X,y){
    m<-melt(X*as.vector(y))
    m<-m[m$value!=0,]
    regions=c()
    medians=c()
    for (region in unique(m$X2)){
        regions<-c(regions,region)
        medians<-c(medians,median(m[m$X2==region,'value']))
    }
    d<-data.frame(cbind(regions,medians))
    return(d)
}
```

#### Precursor Functions

```r
flatten_obs<-function(d){
    # takes in dataframe and flattens it 
    X<-as.matrix(d[(R3names)])
    X_with_other<-as.matrix(d[(R3names)])
    y<-as.matrix(d['Accuracy.50'])
    for (col in seq(ncol(X))){
        X_with_other[X[,col]==1,col]<-y[X[,col]==1]
    }
    m<-melt(X_with_other)
    m<-m[m$value!=0,]
    # save out accuracy 
    accuracy<-m$value
    colnames(m)[2]<-'region'
    colnames(m)[3]<-'accuracy'
    
    # add in info type 
    #X<-as.matrix(d[(R1names)])
    #X_with_other<-as.matrix(d[(R1names)])
    #y<-as.matrix(d['Info.type'])
    #for (col in seq(ncol(X))){
    #    X_with_other[X[,col]==1,col]<-y[X[,col]==1]
    #}
    #m2<-melt(X_with_other)
    #m2<-m2[m2$value!=0,]
    #info.type<-m2$value
    #m<-data.frame(cbind(m,info.type))
    return(m)
}

```

```r
d<-data_50_sig
m<-flatten_obs(d)
head(m)
```

```r
calc_f_from_melt<-function(m){
    regionmeans = c()
    withins = c()
    for (region in R3names){
        accs<-m[m$region==region,'accuracy'] # acc per region
        regionmean<-mean(accs) # region mean
        regionmeans = c(regionmeans,regionmean) # list of region means 
        withinjSS<-sum((accs-mean(accs))**2)
        withins <-c(withins,withinjSS)
    }
    BetweenSS<-sum((regionmeans-mean(regionmeans))**2)/(length(regionmeans)-1)
    WithinSS<-sum(withins)/(nrow(m)-length(regionmeans))
    F = BetweenSS/WithinSS
    return(F)
}
```

```r
calc_d_between_regions<-function(m,region1,region2){
    accs1<-m[m$region==region1,'accuracy']
    accs2<-m[m$region==region2,'accuracy']
    d=mean(accs1)-mean(accs2)
    #print(mean(accs1))
    #print(mean(accs2))
    return(d)
}
```

```r
region1<-'R3Insula'
region2<-'R3Precentral'
calc_d_between_regions(m,region1,region2)
```

```r
single_boot_on_d<-function(d){
    # need d to have PID 
    PIDs<-unique(d$PID) 

    # sample new studies with replacement
    PIDs_boot<-sample(PIDs,size=length(PIDs),replace = TRUE)

    # loop trhough the newly sampled set of studies 
    for (pid in PIDs_boot){
    study_obs<-d[d$PID==pid,] # get the rows for that study 
    study_n<-nrow(study_obs) # get the number 
    study_obs_rows_boot<-sample(seq(study_n),size=study_n,replace=TRUE) # sample from those 
    study_obs_boot<-study_obs[study_obs_rows_boot,]
    d[d$PID==pid,]<-study_obs_boot # put back into dataset  
    }
    return(d)
}
```

```r
dboot<-single_boot_on_d(d)
head(dboot)
```

####  Boot-strapping

```r
d<-data_50_sig[c('Accuracy.50','PID',R3names)]

# calculate true F 
m<-flatten_obs(d) # flatten first 
F_star<-calc_f_from_melt(m) #calculate F 

# calculate true d 
region1<-'R3Insula'
region2<-'R3Precentral'
region1<-'R3Cingulum_Ant'
region2<-'R3Cingulum_Mid'
d_star<-calc_d_between_regions(m,region1,region2)

# assume null - set all region means to be 0 by subtracting off each mean 
dnull<-d
for (region in R3names){
    #print(region)
    dnull[dnull[region]==1,'Accuracy.50']<-d[d[region]==1,'Accuracy.50']-mean(d[d[region]==1,'Accuracy.50'])
    #print(mean(dnull[dnull[region]==1,'Accuracy.50']))
}

# bootstrap to get variability - null distribution 
num_boots=500
F_dist_boot<-c()
d_dist_boot<-c()
m_region1_boot<-c()
m_region2_boot<-c()
for (b in seq(num_boots)){
    dnull_b<-single_boot_on_d(dnull) # single resample from dataset with null enforced 
    m_b<-flatten_obs(dnull_b) # flatten that 
    F_b<-calc_f_from_melt(m_b) # calculate the F-statistic again 
    F_dist_boot<-c(F_dist_boot,F_b) # store F-statistic 
    
    mean1<-mean(m_b[m_b$region==region1,'accuracy'])
    mean2<-mean(m_b[m_b$region==region2,'accuracy'])
    m_region1_boot<-c(m_region1_boot,mean1)
    m_region2_boot<-c(m_region2_boot,mean2)
    
    
    d_dist_boot<-c(d_dist_boot,mean1-mean2)
}
```

```r
median_cl_boot <- function(x, conf = 0.95) {
    lconf <- (1 - conf)/2
    uconf <- 1 - lconf
    require(boot)
    bmedian <- function(x, ind) median(x[ind])
    bt <- boot(x, bmedian, 1000)
    bb <- boot.ci(bt, type = "perc")
    data.frame(y = median(x), ymin = quantile(bt$t, lconf), ymax = quantile(bt$t, 
        uconf))
}
```
