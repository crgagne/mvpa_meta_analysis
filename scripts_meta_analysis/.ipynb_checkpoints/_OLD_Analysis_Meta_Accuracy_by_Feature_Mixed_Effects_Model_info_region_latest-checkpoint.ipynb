```r
model = lme(Accuracy.50.mean~Info.type+Region,random = ~1|PID,data=data_50_mean)
summary(model)
```

Liklihood ratio test, to compare nested models. Fit using ML instead of REML, because its not recommended. 

"Some people use the REML function in place of the usual likelihood to form likelihood ratio tests
and the AIC and BIC criteria. If the test concerns different mean models, this is generally not
recommended, as it is not clear that the restricted likelihood ratio statistic ought to have a

2 distribution when m is large. Thus, it has been advocated to carry out tests involving the
components of  using ML to fit the model. However, if ones main interest is in estimates of the
covariance parameters  (e.g estimating 
2 and  in the AR(1) model), then REML estimators
should be employed because of they are likely to be less biased." ~Ch8 NSCU

```r
model1 = lme(Accuracy.50.mean~Info.type+Region,random = ~1|PID,data=data_50_mean,method='ML')
model2 = lme(Accuracy.50.mean~Region,random = ~1|PID,data=data_50_mean,method='ML')
model3 = lme(Accuracy.50.mean~Info.type,random = ~1|PID,data=data_50_mean,method='ML')
anova(model1,model2)
anova(model1,model3)

```

Wald F test instead. Tests sequentially, but that doesnt matter here. 

```r
anova(model)
```

plot coefficients

```r
s<-summary(model)
coef<-s$tTable[,1] 
se<-s$tTable[,2] 
df<-data.frame(cbind(coef,se))
#print(s$tTable[,0])
df$name<-rownames(df)
df<-df[2:nrow(df),]
```

```r

ggplot(df, aes(y=coef,x=name))+geom_point()+geom_errorbar(aes(ymax = coef + se, ymin=coef- se))+theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

looking at the post-hoc comparisons value has a higher coefficient than all the other types, and Orb has a marginally significant difference from precentral.

```r
options(repr.plot.width=8, repr.plot.height=8)
X = model.matrix(model,data=data_50_mean)
X<-X[,2:ncol(X)]
C = cor(X)

col <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))
col <- colorRampPalette(c("#4477AA", "#77AADD", "#FFFFFF","#EE9988", "#BB4444"))

corrplot(C, method="color", col=col(200),  
         type="upper", order="hclust", 
         addCoef.col = "black", # Add coefficient of correlation
         tl.col="black", tl.srt=90, #Text label color and rotation
         # Combine with significance
         # hide correlation coefficient on the principal diagonal
         diag=FALSE 
         )
```
