```python
## copy the nii file

atlas_img_aal = nib.load('/srv/local/mricron-gtk1-20101203/templates/aal.nii.gz')
coef_img_data = atlas_img_aal.get_data()
coef_img_data2 = coef_img_data.copy().astype('float32')
np.unique(coef_img_data2)
```

```python
## put coefficients into data matrix

intercept = coef.loc[coef.index=='(Intercept)','x'].as_matrix() # get intercept

for r,roi_name in enumerate(labels[1]):
    
    roi_number = labels.loc[r,0] # get roi number for matrix
    
    
    if any(coef.index==roi_name): # check if roi is in my coefficients
        c = coef.loc[coef.index==roi_name,'x'].as_matrix() # coef
        coef_img_data2[coef_img_data==roi_number]=c #(c+intercept)[0]
        #print(roi_name)
        #print(roi_number)
        #print((coef_img_data==roi_number).sum())
    else:
        coef_img_data2[coef_img_data==roi_number]=0### change this..XX 
        
        
## change 
#coef_img_data2[coef_img_data2==0] = intercept; 
```

```python
## save out image

new_img = nib.Nifti1Image(coef_img_data2, atlas_img_affine_aal)

nib.save(new_img, 'coef_model_no_collapse2.nii')
```
